Question 1 : Explain the basic structure of an HTML page. Why is it important to follow this structure?

          The basic structure of an HTML page defines how content is organized and displayed by the browser. A standard HTML document includes the following parts.

```python
<!DOCTYPE html>
<html>
<head>
  <title>My Web Page</title>
</head>
<body>
  <h1>Hello World</h1>
  <p>This is my first webpage.</p>
</body>
</html>
```

     <!DOCTYPE html> tells the browser that this document follows HTML5.

     <html> is the root element that wraps the entire page.

     <head> contains metadata like the page title, character set, and links to CSS.

     <title> defines the text shown in the browser tab.

     <body> contains all the visible content of the page.

     Why it is important:

     1.Ensures the browser correctly understands and renders the page.

     2.Improves consistency across different browsers.

     3.Helps with SEO and accessibility.

     4.Makes the code more readable and maintainable for developers

Question 4: You’re creating a form that asks the user to enter their name, select their country from a list, and click a button to submit. Write the HTML code for this.

```python
<!DOCTYPE html>
<html lang="en">

<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
</head>

<body>
    <form>
        <label for="name">Name:</label><br>
        <input type="text" id="name" name="name"><br><br>

        <label for="country">Country:</label><br>
        <select id="country" name="country">
            <option>India</option>
            <option>USA</option>
            <option>UK</option>
        </select><br><br>

        <button type="submit">Submit</button>
    </form>

</body>

</html>
```

Question 5: Write short HTML code to show your favorite fruits in a bulleted list and your top 3 programming languages in a numbered list.

```python
<!DOCTYPE html>
<html lang="en">

<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
</head>

<body>
    <h3>My favorite fruits</h3>
    <ul>
        <li>Apple</li>
        <li>Mango</li>
        <li>Banana</li>
    </ul>
    <h3>My top 3 programming languages</h3>
    <ol>
        <li>JavaScript</li>
        <li>Python</li>
        <li>C++</li>
    </ol>


</body>

</html>
```
